```python
# Analyze ROAS over time
roas_by_date = (
    df.groupby("date")["roas"]
    .mean()
    .reset_index()
    .sort_values("date")
)

roas_by_date.head(), roas_by_date.tail()
```

```python
mid_point = len(roas_by_date) // 2

early_roas = roas_by_date["roas"].iloc[:mid_point].mean()
late_roas = roas_by_date["roas"].iloc[mid_point:].mean()

early_roas, late_roas
```

```python
summary = {}

# ✅ Build overall metrics safely
summary["overall"] = {}

if "ctr" in df.columns:
    summary["overall"]["avg_ctr"] = df["ctr"].mean()

if "conversion_rate" in df.columns:
    summary["overall"]["avg_conversion_rate"] = df["conversion_rate"].mean()

if "cpc" in df.columns:
    summary["overall"]["avg_cpc"] = df["cpc"].mean()

if "roas" in df.columns:
    summary["overall"]["avg_roas"] = df["roas"].mean()

# ✅ performance by creative type
metrics_for_group = [col for col in ["ctr", "conversion_rate", "roas"] if col in df.columns]

creative_perf = (
    df.groupby("creative_type")[metrics_for_group]
    .mean()
    .reset_index()
)

summary["by_creative_type"] = creative_perf.to_dict(orient="records")

# ✅ performance by audience type
audience_perf = (
    df.groupby("audience_type")[metrics_for_group]
    .mean()
    .reset_index()
)

summary["by_audience_type"] = audience_perf.to_dict(orient="records")

summary
```
